```python
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('/content/BAC.csv')
```

```python
data = {'Category': ['A', 'B', 'C', 'D', 'E'],
        'Value': [25, 35, 15, 20, 10]}
df = pd.DataFrame(data)
```

```python
# 1. Bar Chart
plt.figure(figsize=(8, 6))
plt.bar(df['Category'], df['Value'], color='skyblue')
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Value')
plt.show()
```

```python
# 2. Line Chart
x = np.linspace(0, 10, 100)
y = np.sin(x)
plt.figure(figsize=(8, 6))
plt.plot(x, y, color='red')
plt.title('Line Chart')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 3. Scatter Plot
x = np.random.rand(50)
y = np.random.rand(50)
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='green', alpha=0.7)
plt.title('Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

```python
# 4. Histogram
data = np.random.randn(1000)
plt.figure(figsize=(8, 6))
plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 5. Box Plot
data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.figure(figsize=(8, 6))
plt.boxplot(data, patch_artist=True, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Box Plot')
plt.ylabel('Value')
plt.show()
```

```python
# 6. Pie Chart
sizes = [25, 35, 15, 20, 10]
labels = ['A', 'B', 'C', 'D', 'Other']
plt.figure(figsize=(8, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Pie Chart')
plt.show()
```

```python
# 7. Heatmap
data = np.random.rand(10, 10)
plt.figure(figsize=(8, 6))
sns.heatmap(data, annot=True, cmap='viridis')
plt.title('Heatmap')
plt.show()
```

Graph Types and Use Cases:

1 Bar Chart: Used to compare values across different categories.

2 Line Chart: Used to visualize trends and changes over time.

3 Scatter Plot: Used to show the relationship between two variables.

4 Histogram: Used to display the distribution of a single numerical variable.

5 Box Plot: Used to visualize the distribution of data and identify outliers.

6 Pie Chart: Used to show the proportion of each category in a whole.

7 Heatmap: Used to visualize data in a matrix format, showing the relationship between two variables using color.
